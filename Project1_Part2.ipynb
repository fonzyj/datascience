```python
#  Decide on how to address the missing values and do it! (This requires your judgement, so explain your choice).
# I have used mode to replace the missing values. The resulting mean was not not very far from the initial one prior to replacement.
# Moreover, the mode represents the most common category and can be a reasonable estimate for the missing values.
df.isna().sum()
```

```python
# 6) Confirm that there are no missing values after addressing them.
df['Outlet_Size'] = df['Outlet_Size'].fillna('Medium')
df.isna().sum()
```
